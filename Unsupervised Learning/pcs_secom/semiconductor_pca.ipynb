```python
# filling null values with median values
data.fillna(data.median(), inplace=True)
```

```python
# standarizing values
from sklearn.preprocessing import StandardScaler
standardized_values = StandardScaler().fit_transform(data)
```

```python
standardized_values
```

```python
# transposing values
transpose_values = standardized_values.T
```

```python
Covariance_Matrix = (transpose_values.dot(standardized_values))/len(standardized_values)
```

```python
Covariance_Matrix.shape
```

```python
Covariance_Matrix
```

```python
# calculating eigen values and eigen vectors
Eigen_values, Eigen_Vectors = np.linalg.eig(Covariance_Matrix)
```

```python
index = Eigen_values.argsort()[::-1]   
Eigen_values = Eigen_values[index]
Eigen_Vectors = Eigen_Vectors[:,index]
```

```python
Eigen_Vectors
```

```python
Eigen_values
```

```python
# implementating pca 
from sklearn.decomposition import PCA
pca = PCA().fit(standardized_values)
#Plotting the Cumulative Summation of the Explained Variance
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.grid(True)
plt.ylabel('variance_percentage') #for each component
plt.title('pulsar_data')
plt.show()
```

```python
pca = PCA(n_components=7).fit_transform(standardized_values)
pca =  pd.DataFrame(data = pca, columns = ['pca_0', 'pca_1', 'pca_2', 
                                                                           'pca_3', 'pca_4', 'pca_5', 'pca_6'])
```
