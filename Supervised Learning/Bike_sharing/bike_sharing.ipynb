```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train_data,test_data,train_label,test_label=train_test_split(day.drop(["cnt","instant","yr","mnth","holiday","weekday","dteday"], axis=1),day["cnt"],test_size=0.3,random_state = 23)
```

```python
train_data = train_data.reset_index(drop = True)
train_label = train_label.reset_index(drop = True)
test_data = test_data.reset_index(drop = True)
test_label = test_label.reset_index(drop = True)
```

```python
modelpred=LinearRegression()
modelpred.fit(train_data,train_label)
```

```python
prediction=modelpred.predict(train_data)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
rmse=(np.sqrt(mean_squared_error(train_label,prediction)))
rmse
```

```python
r_squared = r2_score(train_label,prediction)
r_squared
```

```python
prediction1=modelpred.predict(test_data)
```

```python
rmse=(np.sqrt(mean_squared_error(test_label,prediction1)))
rmse
```

```python
r_squared = r2_score(test_label,prediction1)
r_squared
```
